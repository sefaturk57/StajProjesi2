# Task
Calculate the BLEU and ROUGE scores for the model's output against the reference outputs in the dataset.

## Gerekli kütüphaneleri yükle

### Subtask:
BLEU ve ROUGE metriklerini hesaplamak için gerekli kütüphaneleri (NLTK ve rouge_score) yükleyin.


**Reasoning**:
The subtask requires installing NLTK and rouge_score libraries and downloading the 'punkt' resource from NLTK. This can be done in one code block.

## Modelden cevap üret

### Subtask:
Veri setindeki her "instruction" için modelden bir cevap üretin.


**Reasoning**:
Iterate through the dataset, generate responses for each instruction, and store the instruction and generated response.



```
instructions_and_generated_answers = []

for example in veri_seti:
    instruction = example["instruction"]
    giris = f"<|im_start|>user\n{instruction}\n<|im_end|>\n<|im_start|>assistant\n"

    try:
        sonuc = sohbet(giris, max_new_tokens=150, do_sample=True, temperature=0.7)[0]["generated_text"]
        cevap = sonuc.split("<|im_start|>assistant\n")[-1]
        instructions_and_generated_answers.append((instruction, cevap.strip()))
    except Exception as e:
        print(f"Error processing instruction: {instruction}")
        print(f"Error: {e}")
        # Optionally append a placeholder or skip this example
        instructions_and_generated_answers.append((instruction, "Error generating response"))


print(f"Generated responses for {len(instructions_and_generated_answers)} instructions.")
```

## Metrikleri hesapla

### Subtask:
Üretilen cevapları veri setindeki karşılık gelen "output" değerleriyle karşılaştırarak BLEU ve ROUGE skorlarını hesaplayın.


**Reasoning**:
Calculate the BLEU and ROUGE scores for each generated answer against its corresponding reference output and then compute the average scores.



```
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer

reference_outputs = [example["output"] for example in veri_seti]
generated_answers = [answer for instruction, answer in instructions_and_generated_answers]

bleu_scores = []
rouge_scores = {
    "rouge-1": [],
    "rouge-2": [],
    "rouge-l": []
}

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

for ref, gen in zip(reference_outputs, generated_answers):
    # Calculate BLEU score
    # BLEU expects a list of reference sentences, so we wrap the single reference in a list
    # It also expects tokenized sentences
    reference_tokens = [nltk.word_tokenize(ref)]
    generated_tokens = nltk.word_tokenize(gen)
    try:
        bleu = sentence_bleu(reference_tokens, generated_tokens)
        bleu_scores.append(bleu)
    except ZeroDivisionError:
        bleu_scores.append(0) # Append 0 if there's a division by zero (e.g., empty generated text)


    # Calculate ROUGE scores
    scores = scorer.score(ref, gen)
    rouge_scores["rouge-1"].append(scores["rouge1"].fmeasure)
    rouge_scores["rouge-2"].append(scores["rouge2"].fmeasure)
    rouge_scores["rouge-l"].append(scores["rougeL"].fmeasure)

# Calculate average scores
avg_bleu = sum(bleu_scores) / len(bleu_scores) if bleu_scores else 0
avg_rouge_1 = sum(rouge_scores["rouge-1"]) / len(rouge_scores["rouge-1"]) if rouge_scores["rouge-1"] else 0
avg_rouge_2 = sum(rouge_scores["rouge-2"]) / len(rouge_scores["rouge-2"]) if rouge_scores["rouge-2"] else 0
avg_rouge_l = sum(rouge_scores["rouge-l"]) / len(rouge_scores["rouge-l"]) if rouge_scores["rouge-l"] else 0

print(f"Average BLEU score: {avg_bleu}")
print(f"Average ROUGE-1 score: {avg_rouge_1}")
print(f"Average ROUGE-2 score: {avg_rouge_2}")
print(f"Average ROUGE-L score: {avg_rouge_l}")
```
